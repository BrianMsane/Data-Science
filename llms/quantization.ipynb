## Data Types in ML

### Integer Representations
- Unsigned integer (no sign) has a range of n-bits represented by {0, 2^n -1}
    - For example, for 8-bit (torch.uint8): [0, 255]

- Signed integer represent negative and positive numbers using the two's complement representation
    - Range for n-bits is [-2^(n-1), 2^(n-1) -1]
    - For example, 8-bits (torch.int8) has range [-128, 127]


```python
import torch
```

```python
# use iinfo() to get ranges for each data type
print("The usigned data types\n====")
print(torch.iinfo(torch.uint8))


print("\n\nThe signed data types\n====")
print(torch.iinfo(torch.int8))
print(torch.iinfo(torch.int16))
print(torch.iinfo(torch.int32))
print(torch.iinfo(torch.int64))
```

### Floating point representation
3 components:
- Sign: positive or negative (always 1 bit)
- Exponent(range): impact the representable range of the number
- Fraction(precision): impact on the precision of thenumber

FP32, BF16, FP16, FP8 are floating point format with a specific number of bits for the exponent and the fraction. We have normal and subnormal values.

- Floating point 32 is composed of:
    - Sign: 1-bit
    - Exponent (range): 8 bits
    - Fraction (precision): 23 bits
- Floating point 16 is composed of:
    - Sign: 1 bit
    - Exponent: 5 bits
    - Fraction: 10 bits

```python
print(torch.finfo(torch.float16))
print(torch.finfo(torch.bfloat16))
print(torch.finfo(torch.float32))
print(torch.finfo(torch.float64))
```

```python
value = 1/3 # python uses the floating point 64
tensor_fp64 = torch.tensor(value, dtype=torch.float64)
format(value, '.60f') == format(tensor_fp64.item(), '.60f')
```

```python
# Observe precision

print("FP64", format(torch.tensor(value, dtype=torch.float64), '.60f')) # Best precision
print("FP32", format(torch.tensor(value, dtype=torch.float32),'.60f')) # Better precision
print("FP16", format(torch.tensor(value, dtype=torch.float16), '.60f')) # Good precision
print("BF16", format(torch.tensor(value, dtype=torch.bfloat16), '.60f')) # Bad precision
```

Downcasting happens when we convert a higher data type to a lower data type. The value will be converted to the nearest value in the lower data type and this typically result in a loss of information. It is used in Mixed precison training.
- Advantages
    - Reduced memory footprint
    - Increased compute and speed
- Disadvantages
    - Less precise

```python
tensor = torch.rand(1000, dtype=torch.float32)
print("Original Tensor: \n", tensor[:5])
print("Downcasted: \n", tensor.to(torch.bfloat16)[:5])
```

```python
torch.dot(tensor, tensor)
```

```python
torch.dot(tensor.to(torch.bfloat16), tensor.to(torch.bfloat16))
```
