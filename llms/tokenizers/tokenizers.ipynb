```python
from transformers import AutoTokenizer # interact with any tokenizer
```

```python
sentence = "Testing different tokenizers"
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
token_ids = tokenizer(sentence).input_ids
"""
use the tokenizer to process the input sentences and then return token ids.
The token ids represent some specific tokens and to get those tokens, we first need to decode them
"""
print(token_ids)
```

```python
def decode_tokens(ids, tokenizer):
    return tokenizer.decode(ids)
print(decode_tokens(token_ids, tokenizer))
```

```python
print(len(tokenizer)) # Shows that the bert has a vocabulary size of 28996
```

```python
# A list of colors in RGB for representing the tokens
colors = [
    '102;194;165', '252;141;98', '141;160;203',
    '231;138;195', '166;216;84', '255;217;47'
]

def show_tokens(sentence: str, tokenizer_name: str, tokenizer: AutoTokenizer=None):
    """ Show the tokens each separated by a different color """

    # Load the tokenizer and tokenize the input
    if tokenizer is not None:
        tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
    token_ids = tokenizer(sentence).input_ids

    # Extract vocabulary length
    print(f"{tokenizer_name} vocab length: {len(tokenizer)}\n")

    # Print a colored list of tokens
    for idx, t in enumerate(token_ids):
        print(
            f'\x1b[0;30;48;2;{colors[idx % len(colors)]}m' +
            tokenizer.decode(t) +
            '\x1b[0m',
            end=' '
        )
```

```python
text = """
English and CAPITALIZATION
🎵 鸟
show_tokens False None elif == >= else: two tabs:"    " Three tabs: "       "
12.0*50=600
"""
show_tokens(sentence=text, tokenizer_name="bert-base-uncased", tokenizer=tokenizer)
show_tokens(sentence=text, tokenizer_name="bert-base-cased", tokenizer=tokenizer)
# show_tokens(sentence=text, tokenizer_name="gpt-2", tokenizer=tokenizer)
```
