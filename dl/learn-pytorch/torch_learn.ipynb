# Learning Pytorch

```python
import torch
import numpy as np
import warnings

from pprint import pprint

warnings.filterwarnings('ignore')
print(torch.__version__)
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
my_tensor = torch.tensor(
    [
        [1,2,3],
        [4,5,6]
    ],
    dtype=torch.float32,
    device=device,
    requires_grad=True
)

print('the tensor is \n', my_tensor)
print(my_tensor.device) # assess the device in which this tensor is stored
print(my_tensor.shape) #  get the shape of the tensor (expecting a 2 by 3)
print(my_tensor.dtype) # get the data type of the tensor we just created
print(my_tensor.requires_grad) # parameter to see if the tensor requires gradient
```

#### Note
This code creates a 2X3 PyTorch tesnsore with float32 data type, assigns it to a specified device which can eiher be cuda/gpu or cpu, and enables gradient checking for backprogragation. It is important to note that we initialized this tensor with the values we want it to hold. In contrast to this, there are many other methods of Tensor Initialization which are shown below:

- Empty Tensor: Creates an uninitialized tensor with random values
- Random Tensor: generates a random tensor with values between 0 and 1
- Zeros Tensor: creates a tensor filled with zeros
- Ones Tensor: creates a tensor filled with ones
- Arange Tensor: creates a 1 dimensional torch with values upto the specified value and moving on given step
- Linspace Tensor: generate a tensor with evenly separated values
- Normal Distributed Tensor: fills a tensor with values froma  normal or Gaussian distribution using given mean and standard deviation
- Uniform Distributed Tensor: fills a tensor with values from a uniform distribution between 0 and 1
- Diagonal Tensor: creates a tensor with ones in the diagonal line and zeros elsewhere


```python
 # creates an empty tensor of size 3X3 with random values
empty =torch.empty(
    (3, 3),
    device=device,
    requires_grad=True,
    dtype=torch.float16
)

ones = torch.ones(
    (3,3),
    device=device,
    requires_grad=True,
    dtype=torch.float32,
)
zeros = torch.zeros(
    (3,3),
    dtype=torch.double,
    device=device,
    requires_grad=True,
)

x = torch.arange(
    start=0,
    end=5,
    step=1,
    dtype=torch.float64,
    device=device,
    requires_grad=True
)

y = torch.linspace(0.1, 1, 5)

identity = torch.eye(4,4)

normal = torch.empty(1,5).normal_(mean=0, std=1)

uniform = torch.empty(1,5).uniform_(0,1)

pprint(empty)
pprint(ones)
pprint(zeros)
print(x)
print(y)
print(identity)
print(normal)
print(uniform)
```

```python
# DEMONSTRATING TYPE CONVERTION USING ONE TENSOR

tensor = torch.arange(5)
print("converting to Boolean", tensor.bool())
print("Short Tensor", tensor.short())
print("Long Tensor", tensor.long())
print("Half Tensor", tensor.half())
print("Convert to double ", tensor.double())
print("Convert to float ", tensor.float())
```
